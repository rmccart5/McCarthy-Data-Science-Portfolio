```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.graphics.gofplots import qqplot
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.linear_model import LinearRegression

# Load the housing dataset
housing = fetch_california_housing()

# Creates a DataFrame for the features and a Series for the target variable
X = pd.DataFrame(housing.data, columns=housing.feature_names) 
y = pd.Series(housing.target, name='med_house_value')

# Display first 5 rows of the dataset
display(X.head())
display(y.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 10))
for i, col in enumerate(["MedInc", "HouseAge", "AveBedrms", "Population", "AveOccup", "Latitude", "Longitude"]):
    plt.subplot(2, 4, i + 1)
    sns.regplot(x=X[col],
                y=y,
                scatter_kws={'alpha': 0.7},
                line_kws={'color': 'red'})
    plt.title(f"{col} vs. Performance Index")
    plt.xlabel(col)
    plt.ylabel("Performance Index")
plt.tight_layout()
plt.show()
```

```python
# Add a constant term for VIF computation
X_const = sm.add_constant(X)
vif_data = pd.DataFrame()
vif_data["feature"] = X_const.columns
vif_data["VIF"] = [variance_inflation_factor(X_const.values, i)
                   for i in range(X_const.shape[1])]
print(vif_data)
```

```python
# Fit the linear regression model
lin_reg = LinearRegression()
lin_reg.fit(X, y)
y_pred = lin_reg.predict(X)
residuals = y - y_pred
residuals
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_pred, residuals, alpha=0.7)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("Predicted Performance Index")
plt.ylabel("Residuals")
plt.title("Residuals vs. Predicted Values")
plt.show()
```

```python
# Histogram of Residuals
plt.figure(figsize=(8, 6))
plt.hist(residuals, bins=20, edgecolor='black', alpha=0.7)
plt.xlabel("Residuals")
plt.ylabel("Frequency")
plt.title("Histogram of Residuals")
plt.show()
```

```python
# Q-Q Plot of Residuals
plt.figure(figsize=(8, 6))
qqplot(residuals, line='s', ax=plt.gca())
plt.title("Q-Q Plot of Residuals")
plt.show()
```
